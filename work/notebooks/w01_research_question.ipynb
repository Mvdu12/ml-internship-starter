## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Mvdu12/ml-internship-starter/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

print(f"Average sessions per page: {df['sessions_90d'].mean():.2f}")

dead_pages = (df['sessions_90d'] == 0).mean() * 100
print(f"Percentage of pages with zero sessions: {dead_pages:.2f}%")

print(f"Average word count: {df['word_count'].mean():.2f}")
```

from Average sessions per page and Average word count you can notice that
With over 3,000 words per page but very low traffic, the content is too long for the results it gets.
This gap confirms that grouping pages into archetypes is the best way to fix it
